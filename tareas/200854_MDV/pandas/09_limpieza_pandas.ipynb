## Manejo de NaNs o missings

Maneja los datos de tipos missing. Elije una estrategia adecuada dependiendo del tipo de dato que le asignaste a la columna.


Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. ### Imputacion
En este caso lo que vamos a hacer es iterar sobre el df.columns para ir viendo para cada columna que estrategia en particular hacer, dependiendo del tipo de dato.
1-Para los floats, asumiremos que la distribucion de los datos es relativamente simetrica, por lo tanto para cada uno de los nans lo que vamos a hacer es imputar la media de la columna
2-para los ints, en este caso tomaremos en cuenta que los ints son todos aquellas columnas que representan un ID, asi que no tiene sentido en el caso del int, asignar su media ya que solo son, en este caso, identificadores los ints por lo tanto asignamos un missing ya que uno regularmente no trabaja con id's para prediccioness sino para identificar a una cierta tupla o usuario.
3- En los objects y dates usamos missing para no crear ambiguedades ni mentiras en los datos, y tomamos en cuenta solo los datos de los usuarios que si tienen informacion.

```python
# Tu codigo aqui
#primero quitamos todas las columnas que tiene solo nans
df.dropna(axis=1,how='all',inplace=True)
#ahora quitamos las filas que tienen todos los valores nans
df.dropna(axis=0,how='all',inplace=True)

import json

# Lista de columnas para manejar valores missing
columnas_a_imputar = list(df.columns)
# Crear un diccionario para almacenar la informacin de imputacin
informacion_imputacion = {}

# Iterar sobre cada columna y elegir una estrategia de imputacin
for columna in columnas_a_imputar:
    estrategia = None
    valor = None

    # Elegir la estrategia de imputacin segn las caractersticas de la columna
    #usamos mean para sustituir
    if df[columna].dtype == 'float64':
        estrategia = 'media'
        valor = df[columna].mean()
    #explicacion de porque uso missing en int64 esta en el md de arriba
    #pero en los objects, strings e ints metemos missing
    else:
        estrategia = 'identificador'
        valor = 'missing'

    # Realizar la imputacin
    df[columna].fillna(valor, inplace=True)

    # Almacenar la informacin de imputacin en el diccionario
    informacion_imputacion[columna] = {'estrategia': estrategia, 'valor': valor}

```

### Codigo para salvar y cargar JSONs

```python
# Guardar la informacin de imputacin en un archivo JSON
with open('imputacion.json', 'w') as json_file:
    json.dump(informacion_imputacion, json_file)
```
